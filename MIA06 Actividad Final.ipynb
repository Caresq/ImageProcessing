##### Elemento Estructurante Cuadrado

### Filtro de Laplace para Deteccin de Bordes

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Cargar la imagen en escala de grises
image = cv2.imread('coyote03.jpg', cv2.IMREAD_GRAYSCALE)

#Filtro Gausiano para suavizar la imagen
smoothed_image = cv2.GaussianBlur(image, (21, 21), 0)

# Aplicar Laplace
laplacian = cv2.Laplacian(smoothed_image, cv2.CV_64F)

#Normalizacin
laplacian_normalized = cv2.normalize(laplacian, None, 40, 255, cv2.NORM_MINMAX)
laplacian_normalized = np.uint8(laplacian_normalized)

# Aplicacin de un umbral
_, thresholded = cv2.threshold(laplacian_normalized, 130, 150, cv2.THRESH_BINARY)

# Mostrar la imagen original, Laplace, Laplace normalizado y Laplace con Umbral
plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(laplacian, cmap='gray')
plt.title('Imagen con Laplace')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(laplacian_normalized, cmap='gray')
plt.title('Imagen con Laplace Normalizado')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(thresholded, cmap='gray')
plt.title('Imagen con Laplace con Umbral')
plt.axis('off')

plt.show()
```

### Segmentacin por Distancia y Watershed

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Cargar la imagen
image = cv2.imread('coyote03.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Aplicar una umbralizacin binaria
_, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Aplicar apertura morfolgica
kernel = np.ones((3, 3), np.uint8)
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, iterations=2)

# Fondo seguro (sure background)
sure_bg = cv2.dilate(opening, kernel, iterations=3)

# Transformada de distancia para el primer plano seguro (sure foreground)
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
_, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)
sure_fg = np.uint8(sure_fg)

# Regin desconocida
unknown = cv2.subtract(sure_bg, sure_fg)

# Marcadores
_, markers = cv2.connectedComponents(sure_fg)
markers = markers + 1
markers[unknown == 255] = 0

# Aplicar el algoritmo Watershed
markers = cv2.watershed(image, markers)
image[markers == -1] = [255, 0, 0]

# Mostrar los resultados
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original con Contornos')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(binary, cmap='gray')
plt.title('Umbralizacin Binaria')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(markers, cmap='jet')
plt.title('Marcadores Watershed')
plt.axis('off')

plt.tight_layout()
plt.show()
```
